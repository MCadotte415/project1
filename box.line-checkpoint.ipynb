```python
df=pd.read_csv(data)
```

```python
DF['Industry Role'] = DF['Industry'].str.split(',').str[0]
```

```python
DF['Industry Role'].unique()
```

```python
DF['Industry Role'] = DF['Industry Role'].replace({'Accounting ':'Accounting'})
DF['Industry Role'] = DF['Industry Role'].replace({'BPO ':'BPO'})
DF['Industry Role'] = DF['Industry Role'].replace({'IT-Software ':'IT-Software'})
DF['Industry Role'] = DF['Industry Role'].replace({'Banking ':'Banking'})
DF['Industry Role'] = DF['Industry Role'].replace({'Education ':'Education'})
DF['Industry Role'] = DF['Industry Role'].replace({'FMCG ':'FMCG'})
DF['Industry Role'] = DF['Industry Role'].replace({'Pharma ':'Pharma'})
DF['Industry Role'] = DF['Industry Role'].replace({'Strategy ':'Strategy'})
DF['Industry Role'] = DF['Industry Role'].replace({'Telcom ':'Telcom'})
DF['Industry Role'] = DF['Industry Role'].replace({'Telcom':'Telecom '})
```
