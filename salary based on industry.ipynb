```python
import matplotlib.pyplot as plt
import pandas as pd
import scipy.stats as st
import numpy as np
```

```python
data = 'Resources/clean_jobss.csv'
```

```python
df = pd.read_csv(data)
df.head()
```

```python
df['Industry Role'] = df['Industry'].str.split(',').str[0]
```

```python
df['Industry Role 2'] = df['Industry Role'].str.split('/').str[0]
```

```python
df['Industry Role 2'].unique()
```

```python
df['Industry Role 2'] = df['Industry Role 2'].replace({'Accounting ':'Accounting'})
df['Industry Role 2'] = df['Industry Role 2'].replace({'BPO ':'BPO'})
df['Industry Role 2'] = df['Industry Role 2'].replace({'IT-Software ':'IT-Software'})
df['Industry Role 2'] = df['Industry Role 2'].replace({'Banking ':'Banking'})
df['Industry Role 2'] = df['Industry Role 2'].replace({'Education ':'Education'})
df['Industry Role 2'] = df['Industry Role 2'].replace({'FMCG ':'FMCG'})
df['Industry Role 2'] = df['Industry Role 2'].replace({'Pharma ':'Pharma'})
df['Industry Role 2'] = df['Industry Role 2'].replace({'Strategy ':'Strategy'})
df['Industry Role 2'] = df['Industry Role 2'].replace({'Telcom ':'Telcom'})
df['Industry Role 2'] = df['Industry Role 2'].replace({'Telcom':'Telecom '})
```

```python
Accounting_df = df.loc[df["Industry Role 2"] == "Accounting",:]
Banking_df = df.loc[df["Industry Role 2"] == "Banking",:]
Pharma_df = df.loc[df["Industry Role 2"] == "Pharma",:]
Education_df = df.loc[df["Industry Role 2"] == "Education",:]
Advertising_df = df.loc[df["Industry Role 2"] == "Advertising",:]
ITsoftware_df = df.loc[df["Industry Role 2"] == "IT-Software",:]
```

```python
Accounting_max = Accounting_df.groupby('Job Title').max()['sal']
Accounting_vol = pd.DataFrame(Accounting_max)
Accounting_merge = pd.merge(Accounting_vol, df, on=("Job Title","sal"), how="left")

Accounting_jobs = Accounting_merge["sal"]

quartiles = Accounting_jobs.quantile([.25,.5,.75])
lowerq = quartiles[0.25]
upperq = quartiles[0.75]

iqr = upperq-lowerq
lower_bound = lowerq - (1.5*iqr)
upper_bound = upperq + (1.5*iqr)

print(f"The lower quartile of Accounting: {lowerq}")
print(f"The upper quartile of Accounting: {upperq}")
print(f"The interquartile range of Accounting: {iqr}")
print(f"The median of Accounting: {quartiles[0.5]} ")
print(f"Values below {lower_bound} could be outliers.")
print(f"Values above {upper_bound} could be outliers.")
```
